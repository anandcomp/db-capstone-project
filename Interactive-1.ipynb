```python
show_total_paid_customers_over_60 = """SELECT customer.FullName, customer.ContactNumber FROM customer
                    INNER JOIN orders
                    ON customer.CustomerID = orders.Customer_CustomerID
                    WHERE orders.TotalCost > 60
                    """
```

```python
cursor.execute(show_total_paid_customers_over_60)
```

```python
rows = cursor.fetchall()
for row in rows:
    print(f'FullName : {row[0]}, ContactNumber: {row[1]}')
```
